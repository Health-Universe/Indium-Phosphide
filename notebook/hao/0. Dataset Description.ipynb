This notebook describes the dataset used in the project


```python
import numpy as np # linear algebra
import pandas as pd # data processing
from pandas import DataFrame


# Import plotting libraries
import seaborn as sns
import matplotlib 
from matplotlib import pyplot as plt

# Set larger fontsize for all plots
matplotlib.rcParams.update({'font.size': 30})
from IPython.display import clear_output
```

```python
# Getting dataset
df = pd.read_csv('../../dataset/hao_dataset.csv')
df
```

```python
df_aug = pd.read_csv('dataset_augmented.csv')
df_aug
```

### Temperature

```python
temp = pd.unique(df['temp_c'])
print(df['temp_c'].value_counts())
```

```python
a4_dims = (9.8, 6)
fig, ax = plt.subplots(figsize=a4_dims)
ax = sns.violinplot(data=df['temp_c'],inner=None, color='0.85')
ax = sns.swarmplot (data=df['temp_c'], color="orange", edgecolor="gray", size=6.5)
fig.savefig('temp.png', transparent=True)
```

### Diameter

```python
a4_dims = (25, 8)

fig, axs = plt.subplots(ncols=2, figsize=a4_dims)
fig.suptitle('Diameter', fontsize=60)

datasets=(df, df_aug)
i=0
titles=('Initial dataset', 'Final dataset')
for dataset in datasets :

    diameter = dataset.loc[dataset['diameter_nm'] != "None"]
    d=diameter['diameter_nm'].astype(float).tolist()
    sns.violinplot(data=d, inner=None, color='0.9', ax=axs[i]).set(title=titles[i])
    sns.swarmplot (data=d, color="palevioletred", edgecolor="gray", size=9, ax=axs[i])
    i=+1 

fig.savefig('dia.png', transparent=True)
```

### Emission

```python
a4_dims = (25, 8)

fig, axs = plt.subplots(ncols=2, figsize=a4_dims)
fig.suptitle('Emission', fontsize=60)

datasets=(df, df_aug)
i=0
titles=('Initial dataset', 'Final dataset')
for dataset in datasets :

    emission = dataset.loc[dataset['emission_nm'] != "None"]
    e=emission['emission_nm'].astype(float).tolist()
    sns.violinplot(data=e, inner=None, color='0.9', ax=axs[i]).set(title=titles[i])
    sns.swarmplot (data=e, color="steelblue", edgecolor="gray", size=9, ax=axs[i])
    i=+1 

fig.savefig('emission.png', transparent=True)
```

### Absorption

```python
a4_dims = (25, 8)

fig, axs = plt.subplots(ncols=2, figsize=a4_dims)
fig.suptitle('Absorption', fontsize=60)

datasets=(df, df_aug)
i=0
titles=('Initial dataset', 'Final dataset')
for dataset in datasets :

    absorption = dataset.loc[dataset['abs_nm'] != "None"]
    a=absorption['abs_nm'].astype(float).tolist()
    sns.violinplot(data=a, inner=None, color='0.9', ax=axs[i]).set(title=titles[i])
    sns.swarmplot (data=a, color="lightcoral", edgecolor="gray", size=9, ax=axs[i])
    i=+1 
fig.savefig('abs.png', transparent=True)
```
